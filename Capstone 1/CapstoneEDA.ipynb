### 3. Cleaning Data and Checking for Inconsistencies

```python
def perform_eda(df):
    df.set_index('date', inplace=True)
    print("DataFrame Information:")
    print(df.info(),"\n","--"*20)
    print("Missing Values in DataFrame:")
    print(df.isna().sum(),"\n","--"*20)
    print("Duplicate Values in DataFrame:")
    print(df.duplicated().sum(),"\n","--"*20)
    print("Summary Statistics:")
    print(df.describe(),"\n","--"*20)
    
```

```python
inflation = data[data['type'] == 'CPI']

gdp = data[data['type'] == 'GDP']
unemploymentRate = data[data['type'] == 'Unemployment Rate']
bankInterestRate = data[data['type'] == 'Bank Interest Rate']
bondRate = data[data['type'] == 'Bond Rates']
```

### Inflation (CPI)

```python
perform_eda(inflation)
```

```python
def univariate_plot(value, first_title, second_title, third_title , box_plot_variable = None, plot_1_label=None, plot_2_label=None):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))  # Adjust the figsize to fit your needs

    axs[0].hist(value, color='blue', edgecolor='black')  # Data for the first plot
    axs[0].set_title(first_title)
    # axs[0].set_xlabel(plot_1_label)

    axs[1].set_title(second_title)
    axs[1].plot(value)  # Data for the second plot
    # axs[1].set_xlabel(plot_2_label)

    axs[2].boxplot(value)  # Data for the first plot
    axs[2].set_title(third_title)
    axs[2].set_xticklabels([box_plot_variable])

    # Repeat for other subplots as needed
    plt.tight_layout()
    plt.show()
```

```python
univariate_plot(inflation.value, 'Histogram CPI', 'Line Plot CPI', 'Box Plot CPI', 'CPI')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Financial Market:
# Create a pivot table to reshape the data for the correlation matrix

data_fin_f = data_fin[data_fin['date'] > '1992-01-01']
pivot_df_f = data_fin_f.pivot(index='date', columns='type', values='adj close')

# Compute the correlation matrix
correlation_matrix_f = pivot_df_f.corr()

# Set up the matplotlib figure
plt.figure(figsize=(8, 6), facecolor='black')

# Set Seaborn darkgrid style
sns.set_style("darkgrid")

# Create a heatmap using Seaborn
heatmap = sns.heatmap(correlation_matrix_f, annot=True, cmap='coolwarm', linewidths=.5)

plt.title('Correlation Matrix Heatmap', color='white')  # Set the title color to white
plt.xticks(color='white')  # Set x-axis labels color to white
plt.yticks(color='white')  # Set y-axis labels color to white

# Get the color bar (legend) text objects and set their color to white
cbar = heatmap.collections[0].colorbar
cbar.set_label("Legend")
cbar.ax.yaxis.set_tick_params(color="white")
for text in cbar.ax.get_yticklabels():
    text.set_color("white")

# Show the plot
plt.show()
```
